```python
import ast

def load_lists_from_file(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        lines = file.readlines()
        
    # Extract lists from lines
    unique_services = ast.literal_eval(lines[0].split(":", 1)[1].strip())
    unique_active_intents = ast.literal_eval(lines[1].split(":", 1)[1].strip())
    unique_slot_keys = ast.literal_eval(lines[2].split(":", 1)[1].strip())
    
    return list(unique_services), list(unique_active_intents), list(unique_slot_keys)


file_path = 'json_values.txt'
services, intents, slot_keys = load_lists_from_file(file_path)
```

```python
print(intents)
```

```python
from openai import OpenAI


client = OpenAI(
  organization='org_key',
  api_key='api_key'
)
```

```python
def intent_checker(intent):
    # prompt for asserting intents
    prompt = f'please tell me which of these intents: "find_hotel", "book_hotel", "find_train", "book_train"  or "NONE" does "{intent}" match.  Answer in one word'
    

    completion = client.chat.completions.create(
        model="gpt-3.5-turbo-0125",
        max_tokens=40,
        temperature=0,
        messages=[
            {"role": "user", "content": prompt}
        ]
    )
    answer = completion.choices[0].message.content.strip().lower() 

    #print(answer)  # For debugging purposes

    return answer
```

```python
intent_dictionary = {}
for intent in intents:
   result = intent_checker(intent)
   intent_dictionary[intent] = result
```

```python
print(intent_dictionary)
```

```python
# Open a file in write mode
with open("intent_dictionary.txt", "w") as file:
   
    for intent, result in intent_dictionary.items():

        file.write(f"{intent}: {result}\n")
```

```python
def slot_checker(slot_value):
    # Prompt for slot naming enforcement
    prompt = f'please tell me which of these multiwoz2.2 slot values: pricerange, parking, internet, stars, area, type, bookpeople, bookday, bookstay, destination, departure, day, name, address, phone, postcode, ref, arriveby, leaveat, trainid, ref, price, none, duration does "{slot_value}" match.  Name a single match. Answer using a single word only'
    
    completion = client.chat.completions.create(
        model="gpt-3.5-turbo-0125",
        max_tokens=40,
        temperature=0,
        messages=[
            {"role": "user", "content": prompt}
        ]
    )
    answer = completion.choices[0].message.content.strip().lower()  # Ensure answer is in lower case and stripped of whitespace

    #print(answer)  # For debugging purposes

    return answer
```

```python
slot_dictionary = {}
for slot_key in slot_keys:
   result = slot_checker(slot_key)
   slot_dictionary[slot_key] = result
```

```python
with open("slot_dictionary.txt", "w") as file:
    for slot, result in slot_dictionary.items():
        file.write(f"{slot}: {result}\n")
```
